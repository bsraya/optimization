```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_points = np.array([0.5, 0.3, 0.5, 2, 2.5])
x_points = np.array([0.0, 0.5, 1, 1.5, 3.0])
cost = lambda b0, b1: 0.5 * sum((y_points - (b0 + b1 * x_points))**2)
full_cost_arr = []

min, max = -10,13
full_theta = np.arange(min,max)

for i in range(min,max):
    full_cost_arr.append(cost(0,i))

plt.plot(
    full_theta,
    full_cost_arr
)

plt.scatter(
    -5,
    cost(0, -5),
    color = 'red'
)

plt.title("Linear Regression's MSE")
plt.xlabel(r'$\beta_1$')
plt.ylabel(r'$J(0, \beta_1)$')
```

```python
plt.plot(
    full_theta,
    full_cost_arr
)

plt.scatter(
    np.arange(-5, 1),
    full_cost_arr[5:11],
    color = 'red'
)

for j in np.arange(6, 11):
    plt.annotate(
        '', 
        xy=(full_theta[j], full_cost_arr[j]), 
        xytext=(full_theta[j-1], full_cost_arr[j-1]),
        arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
        va='center', 
        ha='center'
    )

plt.title("Gradient Descent with Low Learning Rate")
plt.ylabel(r"$J(0, \beta_1)$")
plt.xlabel(r"$\beta_1$")
```

```python
y_points = np.array([0.5, 0.3, 0.5, 2, 2.5])
x_points = np.array([0.0, 0.5, 1, 1.5, 3.0])
cost = lambda b0, b1: 0.5 * sum((y_points - (b0 + b1 * x_points))**2)
cost_arr = []

theta = [-5, 10, 5, -10, -2, 2]

for i in theta:
    cost_arr.append(cost(0,i))

plt.plot(
    full_theta,
    full_cost_arr
)

plt.scatter(
    theta,
    cost_arr,
    color = 'red'
)

for j in np.arange(1, len(theta)):
    plt.annotate(
        '', 
        xy = (theta[j], cost_arr[j]), 
        xytext = (theta[j-1], cost_arr[j-1]),
        arrowprops = {'arrowstyle': '->', 'color': 'r', 'lw': 1},
        va = 'center', 
        ha = 'center'
    )

plt.title("Gradient Descent with High learning Rate")
plt.ylabel(r"$J(0, \beta_1)$")
plt.xlabel(r"$\beta_1$")
```

```python
x = np.array([0.0, 0.5, 1, 1.5, 3.0])
y = np.array([0.5, 0.3, 0.5, 2, 2.5])

plt.scatter (
    x,
    y
)

plt.plot(
    x, 
    0 - 5 * x,
    color='green'
)

for i in range(len(x)):
    lineXdata = (x[i], x[i]) # same X
    lineYdata = (y[i], 0 - 5 * x[i]) # different Y
    plt.plot(lineXdata, lineYdata, color='red')

plt.title('A Scatter Plot with A Regression Line')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x_points = np.array([0.0, 0.5, 1, 1.5, 3.0])
y_points = np.array([0.5, 0.3, 0.5, 2, 2.5])
x = np.array([0,1,2,3,4])
y = np.array([0,1,2,3,4])

plt.scatter(
    x_points,
    y_points
)

plt.plot(
    x, 
    x - 3,
    color='green'
)

for i in range(len(x)):
    lineXdata = (x_points[i], x_points[i]) # same X
    lineYdata = (y_points[i], -3 + x_points[i]) # different Y
    plt.plot(lineXdata, lineYdata, color='red')

plt.xlabel("Sepal Length")
plt.ylabel("Petal Width")
```
