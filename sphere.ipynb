```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import plotly.graph_objects as go
```

```python
sphere = lambda x, y: x ** 2 + y ** 2
```

```python
x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)
x,y = np.meshgrid(x, y)
z = sphere(x, y)
```

# 1. Drawing The 3D Model with Matplotlib

```python
fig = plt.figure(figsize = (12,7))
ax = fig.add_subplot(projection = '3d')

surface = ax.plot_surface(x, y, z, cmap = cm.viridis, linewidth = 0, antialiased = False)

ax.set_zlim(0,200)
fig.colorbar(surface, shrink = 0.5, aspect = 10)
plt.show()
```

```python
fig = go.Figure(
    data=[
        go.Surface(
            z = z
        ),
        go.Surface(z = z, showscale = False, opacity=0.9),
    ]
)

fig.update_layout(
    scene = dict(
        xaxis = dict(nticks=5, range=[0,40],),
        yaxis = dict(nticks=5, range=[0,40],),
        zaxis = dict(nticks=5, range=[0,200],),
    ),
    width = 700,
    margin = dict(r = 20, l = 10, b = 10, t = 10)
)

fig.show()
```

```python
fig = go.Figure(
  data = go.Contour(
    z = z,
    colorscale='hot',
    line_smoothing = 0.85
  )
)

fig.show()
```
