```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python
himmelblau = lambda x,y: (x**2 + y - 11)**2 + (x + y**2 - 7)**2
```

```python
x = np.linspace(-5,5,200)
y = np.linspace(-5,5,200)
x,y = np.meshgrid(x,y)
z = himmelblau(x,y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x, y, z, rstride=1, cstride=1,
  cmap=cm.winter, linewidth=0.08,
  antialiased=True)
plt.show()
```

```python
fig = go.Figure(
    data=[
        go.Surface(
            z = z
        )
    ]
)

fig.update_layout(
    scene = dict(
        xaxis = dict(nticks=9, range=[0,200],),
        yaxis = dict(nticks=9, range=[0,200],),
        zaxis = dict(nticks=9, range=[0,600],),
    ),
    width = 700,
    margin = dict(r = 20, l = 10, b = 10, t = 10)
)

fig.update_traces(
    contours_z = dict(
        show=True, 
        usecolormap=True,
        highlightcolor="limegreen", 
        project_z=True
    )
)

fig.show()
```

```python
fig = go.Figure(data = go.Contour(
  z = z,
  colorscale='hot',
  line_smoothing = 0.85
))

fig.show()
```
