```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import plotly.graph_objects as go
```

```python
flower = lambda x, y: np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
x = np.linspace(-6, 6, 50)
y = np.linspace(-6, 6, 50)
x, y = np.meshgrid(x, y)
z = flower(x, y)
```

# 1. Plotting The 3D Model in Matplotlib

```python
fig = plt.figure(figsize = (12,7))
ax = plt.axes(projection="3d")
surface = ax.plot_surface(x, y, z, cmap = cm.winter, linewidth = 0, antialiased = False)
plt.show()
```

# 2. Plotting The 3D Model with Plotly

```python
fig = go.Figure(
    data=[
        go.Surface(
            z = z
        )
    ]
)

fig.update_layout(
    scene = dict(
        xaxis = dict(nticks=5, range=[0,50],),
        yaxis = dict(nticks=5, range=[0,50],),
        zaxis = dict(nticks=5, range=[-3,3],),
    ),
    width = 700,
    margin = dict(r = 20, l = 10, b = 10, t = 10)
)

fig.update_traces(
    contours_z = dict(
        show=True, 
        usecolormap=True,
        highlightcolor="limegreen", 
        project_z=True
    )
)

fig.show()
```

```python
fig = go.Figure(
  data = go.Contour(
    z = z,
    colorscale = 'Hot',
    line_smoothing = 0.85
  )
)

fig.show()
```
